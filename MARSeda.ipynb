```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("MARS.csv")
```

```python
df.info()
```

```python
df.head(5)
```

## Descriptive statistics of overall score

```python
print(df['Score:'].describe())
```

## Univariate distribution of overall score

```python
plt.figure(figsize=(9, 8))
sns.distplot(df['Score:'], color='g', bins=100, hist_kws={'alpha': 0.4});
```

## Boxplot of overall score distribution

```python
sns.boxplot(x=df['Score:'])

```

## Correlation values of all questions/columns

```python
df.corr()
```

## Correlation heatmap

```python
plt.figure(figsize=(20,10))
c= df.corr()
sns.heatmap(c, annot=True)
```

## Pairplot of all questions/columns in relation to each other and overall score

```python
sns.set()
sns.pairplot(df, size = 2)
plt.show();
```

## Histogram of overall score distribution

```python
sns.histplot(data=df, x="Score:")
```

## Scatterplot of question that was most positively correlated with overall score (9) I can have a fulfilling and satisfying life)

```python
sns.scatterplot(data=df, x="Score:", y="9) I can have a fulfilling and satisfying life.")
```

## Scatterplot of question that was least positively correlated with overall score (1) I am hopeful about the future.)

```python
sns.scatterplot(data=df, x="Score:", y="1) I am hopeful about the future.")
```

## Scatterplot of questions that were most positively correlated with each other (6) I feel accepted as who I am. and 8) I feel good about myself even when others look down on my illness.)

```python
sns.scatterplot(data=df, x="8) I feel good about myself even when others look down on my illness.", y="6) I feel accepted as who I am.")
```
